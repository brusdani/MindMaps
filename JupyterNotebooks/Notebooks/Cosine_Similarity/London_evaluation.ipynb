```python
import json

# Load the embeddings from the JSON file
with open('london.json', 'r') as f:
    embeddings_data = json.load(f)
```

```python
import numpy as np

# Extract embeddings from the loaded JSON data
embeddings_london = np.array([item['embedding'] for item in embeddings_data])

# Check the shape of the embeddings
print(embeddings_london.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

cos_sim_matrix = cosine_similarity(embeddings_london)


plt.figure(figsize=(8, 6))
sns.heatmap(cos_sim_matrix, cmap='YlGnBu', annot=True, fmt='.2f', xticklabels=[f'London{i+1}.txt' for i in range(10)], yticklabels=[f'London{i+1}.txt' for i in range(10)])
plt.title("Cosine Similarity Matrix: London")
plt.show()
```

```python
import numpy as np

# Assuming temperature1_cos_sim_matrix is your cosine similarity matrix
# Set the diagonal elements to NaN
np.fill_diagonal(cos_sim_matrix, np.nan)

print(cos_sim_matrix)

# Calculate the mean excluding NaN values
mean_cos_sim = np.nanmean(cos_sim_matrix)
print(f"Mean Cosine Similarity (excluding diagonal): {mean_cos_sim:.5f}")
```
